```python
plt.figure(figsize=(15,10))
for i in range(1, len(data.columns)):
    plt.subplot(3,3,i)
    plt.hist(data[data.columns[i]])
    plt.title(data.columns[i])
plt.tight_layout()
plt.show()
```

```python
grouped = data.groupby('clusters')[['temp', 'SO2', 'wind', 'precip','predays','manu','popul']].mean()
grouped
```

```python
sns.barplot(x='clusters',y='temp',data=data)
```

```python
sns.barplot(x='clusters',y='SO2',data=data)
```

```python
sns.barplot(x='clusters',y='wind',data=data)
```

```python
sns.boxplot(x="clusters", y="temp", data=data,palette='rainbow')
```

```python
sns.boxplot(x="clusters", y="SO2", data=data,palette='rainbow')
```

```python
sns.boxplot(x="clusters", y="wind", data=data,palette='rainbow')
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data['SO2'], data['popul'], hue=data['clusters'], palette='rainbow')
plt.xlabel('SO2')
plt.ylabel('Populasi')
plt.title('Scatterplot of Precipitation and Population by Cluster')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=data['temp'], y=data['popul'])
plt.xlabel('Suhu')
plt.ylabel('Populasi')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=data['manu'], y=data['popul'])
plt.xlabel('Manufaktur')
plt.ylabel('Populasi')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data['SO2'], data['temp'], hue=data['clusters'], palette='rainbow')
plt.xlabel('SO2')
plt.ylabel('Suhu (°F)')
plt.title('Scatterplot of Temperature and Precipitation by Cluster')
plt.show()
```

```python
sns.scatterplot(data['wind'], data['popul'], hue=data['clusters'], palette='rainbow')
plt.xlabel('tekanan angin')
plt.ylabel('populasi')
plt.title('Scatterplot of population dan winds by Cluster')
plt.show()
```
